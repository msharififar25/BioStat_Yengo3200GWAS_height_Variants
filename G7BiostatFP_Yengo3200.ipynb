```python
df = pd.read_csv('YENGO_2022_GWAS.txt', sep='\t')
df.head()
```

```python
top_snps = df[df["P"] < 5e-8] #filter GWAS p<5e-8, the top 3200 variants per Yengo article
top_snps.shape
```

```python
top_3200_snps = df.nsmallest(3200, 'P') # lowest p-value = most significant
top_3200_snps.shape
```

```python
top_3200_snps.to_csv('top_3200_snps.csv') #.to_csv is a pandas command to convers to .csv
```
